```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup # for HTML parsing
```

```python
# scrape from webpage into a df
res = requests.get("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
```

```python
# reset type and column names
df = pd.DataFrame(df[0])
df = df.astype('object')
df.rename(columns={0:'PostCode', 1:'Borough',2:'Neighborhood'}, inplace=True)
df.head()
```

```python
# drop first row and rows with no borough assigned
df = df[1:]
df = df[df.Borough != 'Not assigned']
df.head()
```

```python
# If a cell has a borough but a Not assigned neighborhood, 
# then the neighborhood will be the same as the borough
neighbor = df[df.Neighborhood == 'Not assigned']
neighbor
```

```python
# only one row 
df.loc[9,"Neighborhood"] = df.loc[9,'Borough']
```

```python
# group and aggregate to combine neighborhoods for same post code
df.groupby(['PostCode','Borough'], as_index=False, sort=False)
```

```python
# join neighborhoods
df= df.groupby(['PostCode','Borough']).aggregate(lambda x: ', '.join(x))
```

```python
# add index
df = df.reset_index()
df
```

Let's map the neighborhoods in Toronto!

```python
!conda install -c conda-forge folium=0.5.0 --yes 
import folium # map rendering library

print('Libraries imported.')
```

```python
# extract Toronto neighborhoods
df_toronto = df[df.Borough.str.contains('Toronto')]
df_toronto
```

```python
print('Toronto has {} boroughs and {} neighborhoods.'.format(
        len(df_toronto['Borough'].unique()),
        df_toronto.shape[0]
    )
)
```

```python
# Use geopy library to get the latitude and longitude values of Toronto.
!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = 'Toronto, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values and mark the neighborhoods
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
